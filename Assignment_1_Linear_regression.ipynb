```python
# Load the dataset
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
boston = load_boston()
bos = pd.DataFrame(boston.data)
```

```python
print("Dataset_description : ",boston.DESCR)
```

```python
print("Boston_data : ",boston.data)
```

```python
print("Boston_target :",boston.target)
```

```python
print("Features_name of dataset : ",boston.feature_names)
```

```python
# Preparing the dataframe
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df.head()
```

```python
# Adding the dependent or target variables
df["Price"]=boston.target
df.head()
```
